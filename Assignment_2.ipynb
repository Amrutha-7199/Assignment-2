```python
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
data.head()
```

#Task(a)

```python
data.drop(data.columns[0], axis=1, inplace=True) # drop Unnamed column in the data frame
data.head()
```

```python
data.isnull().sum()# check for missing values
```

```python
data.drop(['New_Price'], axis=1, inplace=True)# Drop 'New_Price' column
```

```python
data.shape
```

```python
# Replacing missing values with mode
data['Engine'].fillna(data['Engine'].mode()[0], inplace=True)
data['Power'].fillna(data['Power'].mode()[0], inplace=True)
data['Seats'].fillna(data['Seats'].mode()[0], inplace=True)
data['Mileage'].fillna(data['Mileage'].mode()[0], inplace=True)
data.isnull().sum()
```

#Justification

Here I droped 'New_Price' column where 70% of the values are missing.
I replaced missing values with mode because the count of missing values is not too high. So I replaced with mode.

# Task-B

```python
# Define a function to strip units, checking if the value is a string
def remove_units(value):
    if isinstance(value, str):  # Only process strings
        for unit in [' kmpl', ' km/kg', ' CC', ' bhp', ' lakh']:
            if unit in value:
                return float(value.replace(unit, ''))
    return value  # Return the value as is if it's not a string

# Apply the function to relevant columns
data['Mileage'] = data['Mileage'].apply(remove_units)
data['Engine'] = data['Engine'].apply(remove_units)
data['Power'] = data['Power'].apply(remove_units)

data.head()
```

# Task-C

```python
#data = pd.get_dummies(data, columns=['Fuel_Type', 'Transmission'], drop_first=True)
#processed_data = pd.get_dummies(data, columns=['Fuel_Type', 'Transmission'])

encoded_data = pd.get_dummies(data[['Fuel_Type', 'Transmission']], drop_first=True)
encoded_data = encoded_data.astype(int)
final_data = pd.concat([data, encoded_data], axis=1)
final_data.drop(['Fuel_Type', 'Transmission'], axis=1, inplace=True)

final_data.head()
```

```python
# Filter rows where Car_Age is greater than 5 years
filtered_data = data[data['Car_Age'] > 5]
filtered_data.head()
```

```python
# Rename 'Car_Age' to 'Age_of_Car'
renamed_data = data.rename(columns={'Car_Age': 'Age_of_Car'})
renamed_data.head()
```

```python
# Create a new column for car price depreciation estimate
data['Price_Depreciation'] = data['Price'] * 0.9**data['Car_Age']
data.head()
```

```python
# Arrange data by 'Price' in descending order
arr_data = data.sort_values(by='Price', ascending=False)
arr_data.head()
```

```python
# Group by 'Fuel_Type' and calculate the mean of 'Selling_Price'
sum_data = data.groupby('Fuel_Type').agg(average_selling_price=('Price', 'mean')).reset_index()
sum_data.head()
```
